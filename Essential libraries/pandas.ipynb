**4. Treating Null values**

```python
df.head()
```

```python
# counts the number of missing (NaN) values in each column
df.isnull().sum()
```

```python
# Filling in the Nan Date columns with the most recent date(forward fill)
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
df['Date'] = df['Date'].ffill()
```

```python
# To force all dates into a stanrd YYYY-MM-DD format
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
```

```python
# Filling in the NaN calories with the mean
df['Calories'] = df['Calories'].fillna(df['Calories'].mean())
```

```python
df.isnull().sum()
```

**5. Conditional filtering/slicing**

```python
df.head()
```

```python
# Filtering for rows in which Pulse is above 100
df['Pulse'] = df['Pulse'].astype(int)
high_intensity = df[df['Pulse'] > 100]
print(high_intensity)
```

```python
# Filtering for workouts that lasted over 30 mins and also burned more than 300 calories
Cal_30 = df[(df['Duration'] > 30) & (df['Calories'] > 300)]
print(Cal_30)
```

**6. Data Aggregration and grouping**

```python
# Grouping by Duration to see average calories and pulse for different lengths
df.groupby('Duration') [['Calories','Maxpulse']].mean()
```

```python
df.groupby('Date')[['Pulse','Maxpulse']].max()
```

**7. Creating new features/ Feature engineering**

```python
df['Cal_per_minute'] = df['Calories']/df['Duration']
print(df['Cal_per_minute'])
```

```python
df['Pulse_Spike'] = df['Maxpulse'] - df['Pulse']
print(df['Pulse_Spike'])
```
