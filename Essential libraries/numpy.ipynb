### 1. Array Creation

```python
import array
arr = array.array('i',[1,2,3,4,5])
print(arr)
for i in range(0,3):
    print(arr[i], end=" ")
```

**1.1 Arange method** : Syntax: np.arange(start, step, stop, dtype)

```python
a = np.arange(10).reshape(5,2)
print(a, "\n")

b = np.arange(25, 225, 25).reshape(2,4)
print(b)
```

**1.2 Zero method** : Syntax: np.zeros(number-of-zeros, dtype)

```python
c = np.zeros(4, int).reshape(2,2)
print(c, "\n")

d = np.zeros(7, float).reshape(7,1)
print(d)
```

**1.3 All-one arrays** : np.ones(number-of-ones-in-array, dtype)
<br>
                     np.ones_like(array, dtype)

```python
e = np.ones(9, int).reshape(3,3)
print(e, "\n")

print(arr)
f = np.ones_like(arr, int)#creates an array of given shape and type and a given array with only ones
print(f)
```

**1.4 Linspace method** : np.linspace(start, stop, number-of-digits-in-array, dtype, restep=T/F)

```python
g = np.linspace(100, 200, 10, retstep=False, dtype=int)
print(g, "\n")

h = np.linspace(10,20,10, dtype=int).reshape(5,2)
print(h)
```

**1.5 Identity matrix** 

```python
i = np.eye(3, dtype=int)
print(i, "\n")

j = np.eye(9, dtype=complex)
print(j)
```

**1.6 Empty & full array** 

```python
k = np.empty((5,1), dtype=int)
print(k, "\n")

l = np.full([4,4], 10, dtype=int)
print(l)
```

### 2. Array Manipulation

**2.1 Copy and view array**

```python
x = np.array([10,20,30,40])
print(x)

v = x.view()
x[1]=90
print(v)
print(x)

#both their memory locations i.e. IDs are the same
print(id(x), id(v))
```

```python
y = np.array(x)
print(y)

c=y.copy()
print(c)

c[1]=20
print(y)#doesnt change when copy is changed
print(c)

print(id(y), id(c))#different memory locations
```

**2.2 Appending values at the end**

```python
a = np.array([[1,2,3],[4,5,6]]).reshape(2,3)
print(a)

append_col = np.arange(7,13).reshape(2,3)
print(append_col, "\n")

a_col = np.append(a, append_col, axis=0)
print(a_col, "\n")

append_row = np.arange(7,10).reshape(3,1)
print(append_row)

a_row = np.append(a, append_row)
print(a_row)
```

**2.3 Concat index vice**

```python
a = np.array([3,4,5])
b = np.array([9,10,39])
c = np.hstack((a,b))
print(c)

d = np.dstack((a,b))
print(d)
```

**2.4 Split arrays**

```python
print(np.split(c,3))
```

**2.5 Compare arrays**

```python
d = d.reshape(1,6)
e = c == d
print(e)
```

**2.6 Union of 1D arrays**

```python
print(c,d)
e = np.union1d(c,d)
print(e)
```

**2.7 Trimming zeroes**

```python
array = np.array([2,3,4,5,0,0,0,0])
f = np.trim_zeros(array)
print(f)
```

### 3. Matrix functions in NumPy

```python
x = np.array([[1, 2], [4, 5]]) 
y = np.array([[7, 8], [9, 10]]) 
# +,-,* and / can be done

z = np.dot(x,y)
print(z)

z = np.sqrt(x)
print(z)

z = np.sum(y,axis=0)
print(z)

z = x.T
print(z)
```
