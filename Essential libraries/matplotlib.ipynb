```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('data.csv')
df.head()
```

**Notes:**
<br>
1. Column containing time-based data always goes on the X-axis
2. X-axis variable is the independant variable/ the 'cause' or the input you control
3. Y-axis variable is the dependant one/ the 'effect' or the outcome you are measuring


**1. Line Graph**

```python
# Filling in the Nan Date columns with the most recent date(forward fill)
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
df['Date'] = df['Date'].ffill()
# To force all dates into a stanrd YYYY-MM-DD format
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
```

```python
plt.plot(df['Date'], df['Pulse'])
plt.tick_params(axis = 'x', labelsize=7)

plt.show()
```

**2. Scatterplot**

```python
plt.title("Pulse and Maxpulse scatter plot")

plt.scatter(df['Pulse'], df['Maxpulse'])
plt.tight_layout()
plt.xlabel("Pulse")
plt.ylabel("Maxpulse")
plt.grid(True)
plt.show()
```

```python
plt.title("Duration and Calories scatterplot")
plt.scatter(df['Duration'], df['Calories'])
plt.show()
```

```python
plt.title("Duration histogram")
plt.hist(df['Duration'])
plt.show()
```

**4. Multiple line graph**

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Pulse'], label='Average Pulse', color='blue')
plt.plot(df['Date'], df['Maxpulse'], label='Peak Pulse', color='orange', linestyle='--')
plt.title("Comparing pulse and maxpulse")
plt.xlabel('Date')
plt.ylabel('Heart Rate (BPM)')
plt.legend()
plt.show()
```
